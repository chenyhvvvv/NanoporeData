```python
from utils.normalization import normalize_signal_wrapper
import matplotlib.pyplot as plt
def count_continuous_elements(lst):
    count = 1
    result = []
    for i in range(1, len(lst)):
        if lst[i] == lst[i-1]:
            count += 1
        else:
            result.append(count)
            count = 1
            if i == len(lst)-1:
                result.append(count)
    if not result:
        result.append(len(lst))
    return result

def get_segmentation(read_name):
    # read_name = summary[summary["read_index"] == read_index].read_name.item()
    # i = keys_list.index(read_name)
    k = read_name
    sample = normalize_signal_wrapper(visualize_reads[k].raw, visualize_reads[k].offset, visualize_reads[k].range,
                                     visualize_reads[k].digitisation,
                                      method='noisiest', samples=100, threshold=6.0, factor=1.4826)
    sub_df = df[df['read_index'].astype("int") == int(summary[summary["read_name"] == read_name].read_index)]
    sub_df["start_idx"] = sub_df["start_idx"].astype("int")
    sub_df["end_idx"] = sub_df["end_idx"].astype("int")
    sub_df = sub_df.sort_values(by=['start_idx'])
    pos_1 = min(sub_df['start_idx'])
	# If there is a long repeated sequence in the beginning, then recognize it
    repeat_list = count_continuous_elements(sub_df.sort_values(by=['start_idx'])["reference_kmer"].to_numpy())
    # Check if it is some repeat on the beginning of the alignment
    idx_count = 0
    pos_2 = None
    repeat_str = None
    for i in range(10):
        idx_count += repeat_list[i]
        if repeat_list[i] > 25:
            pos_2 = idx_count
            repeat_str = sub_df["reference_kmer"].iloc[i]
    if pos_2:
        return read_name, sub_df, sample, sub_df["start_idx"].iloc[pos_2], repeat_str, repeat_list
    else:
        return read_name, sub_df, sample, pos_1, "XXXXX", repeat_list
```

```python
pd.set_option('mode.chained_assignment', None)
import ruptures as rpt

read_name_list = []
align_df_list = []
align_repeat_list = []
bar = range(len(path_list))
for i in bar:
    try:
        r_n, sub_df, sample, pos, re, re_l = get_segmentation(path_list[i][1])
        read_name_list.append(r_n)
        align_df_list.append(sub_df)
        align_repeat_list.append(re_l)
        # if re != "AAAAA":
        #     continue
        n_bkps = 1
        algo = rpt.Dynp(model="l2", jump=100).fit(sample)
        result = algo.predict(n_bkps=n_bkps)
        result = [result[0], pos, result[1]]
        fig, ax = rpt.display(sample, result, result)
        re = str(len(align_repeat_list) - 1)
        if isinstance(ax, list):
            ax[0].set_title(re)
        else:
            ax.set_title(re)
    except Exception as e:
        pass
```
